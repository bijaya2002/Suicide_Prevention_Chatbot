```python
!pip install flask-ngrok
!pip install nltk

from flask import Flask, render_template , request
from flask_ngrok import run_with_ngrok
import nltk
from nltk.chat.util import Chat, reflections


nltk.download('punkt')

paris = [
    [
        r"(.*)suicide(.*)",
        ["I am really sorry you are feeling this way . Please reach out to a helpline immediately.Here is a helpline number for your country : %s" % '1-008-273-8255']
    ],
    [
        r"I feel(.*)",
        ["Why do you feel %1?", "What makes you feel %1?"]
    ],
    [
        r"(.*)help(.*)",
        ["Sure, I'm here to help. Please tell me more about your situation"]
    ],
    [
        r"quite",
        ["Bye. Take Care. Remember, help is always available."]
    ],
    [
        r"(.*)",
        ["I am here to listen.Please tell me more."]
    ]
]
reflection = {
    "i am": "you are",
    "i was": "you were",
    "i": "you",
    "i'm": "you are",
    "i'd": "you would",
    "i've": "you have",
    "i'll": "you will",
    "my": "your",
    "you are": "I am",
    "you were": "I was",
    "you've": "I have",
    "you'll": "I will",
    "your": "my",
    "yours": "mine",
    "you": "me",
    "me": "you"

}
chat = Chat(paris, reflections)

app = Flask(__name__)
run_with_ngrok(app)

@app.route("/")
def home():
  return render_template("index.html")

@app.route("/get")
def get_bot_response():
  user_text = request.args.get('msg')
  response = chat.respond(user_text)
  if "suicide" in user_text.lower():
    response += "please call a helpline immediately."
  return response

from IPython.display import HTML


html_template = """
<!Doctype html>
<html>
<head>
<title>Suicide Prevention Chatbot</title>
<script>
function getResponse() {
    var userText = document.getElementById("userInput").value;
    var xhr = new XMLHttpRequest();
    xhr.open("GET", "/get?msg=" + userText, true);
    xhr.onland = function() {
        if (xhr.status === 200) {
            document.getElementById("botResponse").innerHTML = xhr.responseText;

        }
    }
    xhr.send();
}
</script>
</head>
<body>
<h1>Suicide Prevention Chatbot</h1>
<div>
     <input id="userInput" type="text" placeholder="Type your message here..." onekeydown="if(event.keyCode ===13) getBotResponse()">
     <button onclick="getResponse()">Send</button>
     </div>
     <div id="botResponse"></div>
     </body>
     </html>
     """

import os
if not os.path.exists('templates'):
    os.makedirs('templates')
with open('templates/index.html', 'w') as file:
     file.write(html_template)
HTML(html_template)

if __name__ =="__main__":
  app.run()
```
